```python
from fabric import Connection, ThreadingGroup, exceptions
from fabric.exceptions import GroupException
```

```python
user = 'jetson'
host = '147.47.200.209'
ports = [str(20100+rack*10+device) for rack in range(0,4) for device in range(1,7)]
```

```python
connections = [
    Connection( f'{user}@{host}:{port}',connect_kwargs={"password": 'jetson'}) for port in ports
]
group = ThreadingGroup.from_connections(connections)
```

```python
while True:
    command = input()
    try:
        clear_output(wait=True)
    except:
        pass
    try:
        results = group.run(command, hide=True)
        for r in results:
            print(f'SUCCESS @ {r.port}' )
            print('---------------------------------------------------------------------------------------')
            print(f'{results[r].stdout.strip()}')
            print('---------------------------------------------------------------------------------------\n')
    except GroupException as e:
        for c, r in e.result.items():
            print(f'FAIL @ {c.port}')
            break
```
